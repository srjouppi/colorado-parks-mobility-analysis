```python

import pandas as pd
```

```python
# Importing daily reports for Denver County air quality from EPA's Air Quality Index Daily Reports
df_2020 = pd.read_csv('aqidaily2020.csv')
df_2021 = pd.read_csv('aqidaily2021.csv')
```

```python
df_2020
```

```python
# Combining the two dataframes
df = pd.concat([df_2020, df_2021], ignore_index=True)
```

```python
df
```

```python
# Applying Air Quality Index ranking to AQI Value
def ranking(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Moderate'
    elif aqi <= 150:
        return 'Unhealthy for Sensitive Groups'
    elif aqi <= 200:
        return 'Unhealthy'
    elif aqi <= 250: 
        return 'Very Unhealthy'
    else:
        return 'Hazardous'
```

```python
df['ranking'] = df['Overall AQI Value'].apply(ranking)
df
```

```python
df['datetime'] = pd.to_datetime(df.Date, format='%m/%y/%Y')
```

```python
df['year'] = df.datetime.dt.year
```

```python
df['month'] = df.datetime.dt.month
```

```python
aq_rank = df.groupby(['year','month']).ranking.value_counts().to_frame().rename({'ranking': 'rank_count'}, axis=1).reset_index()
```

```python
aq_rank['date'] = 15
```

```python
def month_name (month):
    if month == 1:
        return 'January'
    if month == 2:
        return 'February'
    if month == 3:
        return 'March'
    if month == 4:
        return 'April'
    if month == 5:
        return 'May'
    if month == 6:
        return 'June'
    if month == 7:
        return 'July'
    if month == 8:
        return 'August'
    if month == 9:
        return 'September'
    if month == 10:
        return 'October'
    if month == 11:
        return 'November'
    if month == 12:
        return 'December'
```

```python
aq_rank['month_name'] = aq_rank.month.apply(month_name)
```

```python
aq_rank
```

```python
aq_rank_pivot_21 = aq_rank.query('year == 2021').drop(columns={'date','year','month'},axis=1).pivot(index='month_name', columns='ranking').fillna(0).reset_index()
```

```python
aq_rank_pivot_21.to_csv("denver-county-aqi-monthly-totals-2021.csv", index=False)
```

```python
aq_rank_pivot_20 = aq_rank.query('year == 2020').drop(columns={'date','year','month'},axis=1).pivot(index='month_name', columns='ranking').fillna(0).reset_index()
```

```python
aq_rank_pivot_20.to_csv("denver-county-aqi-monthly-totals-2020.csv", index=False")
```
